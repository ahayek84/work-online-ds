# While 

```python
i = 1
while i <= 5:
    print(i)
    i += 1 
```

```python
i = 5
while i > 0:
    print(i)
    i -= 1 
```

# Break & Continue 

```python
while True:
    name = input('what is your name? ')
    if name.startswith('#'):
        continue # go to line 1 
    print('hello ' + name)
    ans = input('do it again? y/n')
    if ans == 'n':
        print('good bye')
        break # go to line 10 
print('done')
```

generate random list with 10000 elements, ask the user to input a number and search for this number. Note: search only in even numbers 

```python
import random
list1 = [random.randint(1, 100) for _ in range(10)]
```

```python
list1
```

```python
big_list = [random.randint(1, 100) for _ in range(10000)]
big_list[:10]
```

```python
while True:
    number = int(input('please enter a positive int even number: '))
    if number % 2 == 0:
        break
print(number)
```

```python
found = False
for num in big_list:
    if num % 2 != 0:
        continue 
    if num == number:
        found = True 
        break 
if found:
    print('found')
else:
    print('not found')
```
